# Wetbulb calculation

```python
## Wetbulb calculation using Stull formula,https://www.omnicalculator.com/physics/wet-bulb
import math
import pandas as pd 
df=pd.read_csv('5year_Weather_EIA_59city_wointerpolated_withtag.csv') 
T=df['TEMPERATURE (F)']
RH=df['HUMIDITY (%)']
result=[]
for i in range(len(T)):
    temp=(convert_fahrenheit_to_celsius(T[i])*math.atan(0.151977*(math.sqrt(RH[i]+8.313659)))+0.00391838*(math.sqrt(RH[i]*RH[i]*RH[i]))*math.atan(.023101*RH[i])-math.atan(RH[i]-1.677331)+math.atan(convert_fahrenheit_to_celsius(T[i])+RH[i])-4.686035)
    result.append(convert_celsius_to_fahrenheit(temp))
    #result.append(T[i]*math.atan(0.151977*(math.sqrt(RH[i]+8.313659)))+0.00391838*(math.sqrt(RH[i]*RH[i]*RH[i]))*math.atan(.023101*RH[i])-math.atan(RH[i]-1.677331)+math.atan(T[i]+RH[i])-4.686035)
df['WETBULB TEMPERATURE (F)']=result
```

# Offsite water calculation

```python
## Offsite water using the fuel coefficients. It should be text. it should be in a csv. Describe in a details.
## Change the code
## Change it to vector operations
import pandas as pd 
from datetime import datetime, timedelta 
import numpy as np
import math
coal_fac=.48*3.78541
hydro_fac=5.99*3.78541
gas_fac=.21*3.78541
nuclear_fac=.61*3.78541
other_fac=.2*3.78541
petro_fac=.36*3.78541
solar_fac=0*3.78541
wind_fac=0*3.78541

df=pd.read_csv("5year_Weather_EIA_59city_wointerpolated_withtag.csv") 
T=df
result=[]
for val in range(len(T)):
    coal=T["COAL"][val]*1000*coal_fac
    hydro=T["HYDRO"][val]*1000*hydro_fac
    gas=T["NATURAL GAS"][val]*1000*gas_fac
    nuclear=T["NUCLEAR"][val]*1000*nuclear_fac
    other=T["OTHER"][val]*1000*other_fac
    petro=T["PETROLEUM"][val]*1000*petro_fac
    solar=T["SOLAR"][val]*1000*solar_fac
    wind=T["WIND"][val]*1000*wind_fac
    
    sum_all=coal+hydro+gas+nuclear+other+petro+solar+wind
    power=T["COAL"][val]*1000+T["HYDRO"][val]*1000+T["NATURAL GAS"][val]*1000+T["NUCLEAR"][val]*1000+T["OTHER"][val]*1000+T["PETROLEUM"][val]*1000+T["SOLAR"][val]*1000+T["WIND"][val]*1000
    if power==0:
        result.append(result[val-1])
    else:
        result.append(sum_all/power)
```
